```python
x,y = new_df.shape
l1 = []
l2 = []
for i in  range(x):
    l1.append(new_df["Marks_obtained"][i])
    l2.append(new_df["Name_of_the_student"][i])
n = len(l1)
for i in range(n):
    for j in range(n-i-1):
        if l1[j+1] > l1[j]:
            l1[j],l1[j+1] = l1[j+1],l1[j]
            l2[j],l2[j+1] = l2[j+1],l2[j]
number = int(input("enter up to how many number of students"))
for i in range(number):
    print(l1[i],l2[i])
        
    
    
    

```

```python
meand = new_df['Marks_obtained'].mean()
stdd = new_df['Marks_obtained'].std()
print(meand,stdd)
list_grade = []
x,y = new_df.shape
for i in range(x):
    grade = ""
    if new_df['Marks_obtained'][i] >= meand + (1.5 * stdd):
        grade = "O"
    elif (meand + 1.0 * stdd) <= new_df['Marks_obtained'][i] < meand+ (1.5 * stdd):
        grade = "A+"
    elif (meand + 0.5 * stdd) <= new_df['Marks_obtained'][i] < meand + (1.0 * stdd):
        grade = "A"
    elif (meand - 0.5 * stdd) <= new_df['Marks_obtained'][i] < meand + (0.5 * stdd):
        grade = "B+"
    elif (meand - 1.0 * stdd) <= new_df['Marks_obtained'][i] < meand - (0.5 * stdd):
        grade = "B"
    elif (meand - 1.5 * stdd) <= new_df['Marks_obtained'][i] < meand - (1.0 * stdd):
        grade = "C"
    elif 35 <= new_df['Marks_obtained'][i] < meand - (1.5 * stdd):
        grade = "P"
    else:
        grade = "Fail"
    list_grade.append(grade)
print(list_grade)
    
```

```python
new_df["Grade"] = list_grade
new_df
```
